#### 📐 DataFrame.apply() [In Series]

```python
df['email'].apply(len)
```

```python
def update_email(email):
   return email.upper()

# The apply method to a function (that change email to all upper cases)
df['email'] = df['email'].apply(update_email)
df
```

```python
# The lambda method (that changes email back to lower case)
df['email'] = df['email'].apply(lambda x: x.lower())
df
```

```python
df['email'].apply(len)
```

```python
# Len function to each series data (num of rows)
df.apply(len)
```

```python
# Len function to each columns instead
df.apply(len, axis='columns')
```

```python
len(df['email'])
```

```python
# Get the Series minimum from the rows
df.apply(pd.Series.min)
```

```python
# The lambda method to get the Series minimum from the rows
df.apply(lambda x: x.min())
```

#### 📐 DataFrame.applymap() [Only work on DataFrames]

#### 📐 DataFrame.map() [Only work on Series]

```python
# The .map() is used for substituting each value in a Series with another value. However, any data that is not within the condition will become an NaN data type.
df['first'].map({'Corey': 'Chris', 'Jane': 'Mary'})
```

```python
# Use .replace() perserves the old data that's not within the condition.
df['first'] = df['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'})
df
```

```python
# What % of people from each country know Python?

# Total number of respondent by each country
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# Concat Series
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
